```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('score_data.csv')
```

```python
# Compute win, tie, and loss counts
wins = (df["Score1"] > df["Score2"]).sum()
ties = (df["Score1"] == df["Score2"]).sum()
losses = (df["Score1"] < df["Score2"]).sum()

# Compute percentages
total_games = len(df)
win_percentage = wins / total_games * 100
tie_percentage = ties / total_games * 100
lose_percentage = losses / total_games * 100

# Print results
print(f"Win Percentage: {win_percentage:.2f}%")
print(f"Tie Percentage: {tie_percentage:.2f}%")
print(f"Lose Percentage: {lose_percentage:.2f}%")
```

```python
rolling_mean_1 = df['Score1'].rolling(window=50, min_periods=1).mean()
rolling_mean_2 = df['Score2'].rolling(window=50, min_periods=1).mean()

plt.plot(rolling_mean_1, color = 'b', label = 'Biased')
plt.plot(rolling_mean_2, color = 'r', label = 'Trained')
plt.legend()
```

```python
# Assuming df contains columns 'Score1' and 'Score2'
score_diff = df['Score1'] - df['Score2']

# Compute rolling mean (e.g., 50-episode window)
rolling_mean = score_diff.rolling(window=50, min_periods=1).mean()

# Plot original data
plt.plot(score_diff, color='b', alpha=0.3, label='Raw Score Difference')

# Plot rolling mean
plt.plot(rolling_mean, color='r', linewidth=2, label='Rolling Mean (50 episodes)')

# Labels and legend
plt.xlabel('Episode')
plt.ylabel('Score Difference (Player1 - Player2)')
plt.title('DQN Training Progress')
plt.legend()
plt.show()
```

```python
# Assuming df contains columns 'Score1' and 'Score2'
reward = df['Reward']

# Compute rolling mean (e.g., 50-episode window)
rolling_mean = reward.rolling(window=50, min_periods=1).mean()

# Plot original data
plt.plot(df['Reward'], color='b', alpha=0.3, label='Reward')

# Plot rolling mean
plt.plot(rolling_mean, color='r', linewidth=2, label='Rolling Mean (50 episodes)')

# Labels and legend
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.title('DQN Training Progress')
plt.legend()
plt.show()
```
