```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("epas/max_epas.csv")
```

```python
df
```

```python
df['Yardline_str'] = df['Yardline'].astype(str)  # Convert Yardline to string

dff = df[((df['Yardline'] > 10) & (df['DownAndDistance'] == "1-10")) | 
         ((df['Yardline'] <= 10) & (df['DownAndDistance'] == "1-" + df['Yardline_str']))
         & (df['Yardline'] != 0)]
```

```python
plt.scatter(dff['Yardline'], dff['Expected Points Added (EPA)'], alpha=0.6)
plt.xlabel("1st Down Yardline")
plt.ylabel("Expected Points Added (EPA)")
plt.title("EPA vs. 1st Down Yardline")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
df['Yardline_str'] = df['Yardline'].astype(str)  # Convert Yardline to string

dfs = df[((df['Yardline'] > 10) & (df['DownAndDistance'] == "2-10")) | 
         ((df['Yardline'] <= 10) & (df['DownAndDistance'] == "2-" + df['Yardline_str']))
         & (df['Yardline'] != 0)]
```

```python
plt.scatter(dfs['Yardline'], dfs['Expected Points Added (EPA)'], alpha=0.6)
plt.xlabel("2nd Down Yardline")
plt.ylabel("Expected Points Added (EPA)")
plt.title("EPA vs. 2nd Down Yardline")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
plt.scatter(dff['Yardline'], dff['Expected Points Added (EPA)'], alpha=0.6)
plt.scatter(dfs['Yardline'], dfs['Expected Points Added (EPA)'], alpha=0.6, color='r')
plt.xlabel("Yardline")
plt.ylabel("Expected Points Added (EPA)")
plt.title("EPA vs. Yardline")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
df['Yardline_str'] = df['Yardline'].astype(str)  # Convert Yardline to string

dft = df[((df['Yardline'] > 10) & (df['DownAndDistance'] == "3-10")) | 
         ((df['Yardline'] <= 10) & (df['DownAndDistance'] == "3-" + df['Yardline_str']))
         & (df['Yardline'] != 0)]
```

```python
plt.scatter(dff['Yardline'], dff['Expected Points Added (EPA)'], alpha=0.6)
plt.scatter(dfs['Yardline'], dfs['Expected Points Added (EPA)'], alpha=0.6, color='r')
plt.scatter(dft['Yardline'], dfs['Expected Points Added (EPA)'], alpha=0.6, color='g')

plt.xlabel("Yardline")
plt.ylabel("Expected Points Added (EPA)")
plt.title("EPA vs. Yardline")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
